```python
def update_col(df1: pd.DataFrame, df2: pd.DataFrame, join_on: str, target_col: str, reference_col: str):
    _cols = df1.columns
    _temp = pd.merge(df1, df2, how='left', on=join_on, suffixes=(None, '_merged'))

    if reference_col in _cols:
        _reference = f"{reference_col}_merged"
    else:
        _reference = reference_col

    _temp[target_col] = _temp[_reference]

    return _temp[_cols]
```

```python
def parse_id(text: str, index: int, splitter: str = '-'):
    try:
        return text.split(splitter)[index]
    except IndexError:
        return None
```
